## Step 1: Build a Seat

Create a class called `Seat` with two attributes:

- `free` which is a boolean.
- `occupant` which is a string.

and 2 methods (functions): 

- `set_occupant(name)` which allows the program to assign someone a seat if it's free
- `remove_occupant()` which  remove someone from a seat and return the name of the person occupying the seat before


```python
class Seat:
    def __init__(self, free=True, occupant=""):
        self.free = free
        self.occupant = occupant

    def set_occupant(self, name):
        if not self.free:
            return False
        self.occupant = name
        self.free = False
        return True

    def remove_occupant(self):
        previous = self.occupant
        self.occupant = ""
        self.free = True
        return previous
```

```python
# Test your code (assign yourself you a Seat)
my_seat = Seat()
print("Free before:", my_seat.free)

my_seat.set_occupant("Stephane")
print("Occupant:", my_seat.occupant, "| Free:", my_seat.free)

# Try to assign to an already-taken seat
assigned = my_seat.set_occupant("Anna")
print("Could reassign?", assigned)

# Remove and check return value
previous = my_seat.remove_occupant()
print("Removed:", previous, "| Free now:", my_seat.free)
```

## Step 2: Build a Table

Create a class `Table` with ? attributes:

- `capacity` which is an integer
- `seats` which is a list of `Seat` objects (size = `capacity`)

and 3 methods (functions): 
- `has_free_spot()` that returns a boolean (True if a spot is available)
- `assign_seat(name)` that places someone at the table
- `left_capacity()` that returns an integer

Question: Which attributes make sense to give? For now let's say we want to build 6 tables with 4 seats.


```python
class Table:
    def __init__(self, capacity=4, seats=None):
        self.capacity = capacity
        if seats is None:
            self.seats = [Seat() for _ in range(capacity)]
        else:
            self.seats = seats
    def has_free_spot(self):
        return any(seat.free for seat in self.seats)
    def assign_seat(self, name):
        for seat in self.seats:
            if seat.free:
                return seat.set_occupant(name)
        return False
    def left_capacity(self):
        return sum(seat.free for seat in self.seats)
```

```python
# Test your code (assign the colleagues at your table to a Table)
table=Table()
print(table.left_capacity())
table.assign_seat("Stephane")
table.assign_seat("Gaetan")
table.assign_seat("Gunay")
print(table.has_free_spot())
table.assign_seat("Sitara")
for s in table.seats:
    print(s.occupant) 
print(table.has_free_spot())
print(table.left_capacity())
 
 
```

## Step 3: Build an OpenSpace

Create a class `Openspace` that contains these attributes:

- `tables` which is a list of `Table`.
- `number_of_tables` which is an integer.

And some methods:

- `organize(names)` that will:
  - **randomly** assign people to `Seat` objects in the different `Table` objects. (hint: explore the `random` library )
- `display()` display the different tables and there occupants in a nice and readable way
- `store(filename)` store the repartition in an file

```python
import random

class Openspace:
    def __init__(self, tables=None, number_of_tables=0):
        if tables is None:
            self.tables = [Table() for _ in range(number_of_tables)]
        else:
            self.tables = tables
    def organize(self, names):
        seats=[]    
        i=0
        while i < len(self.tables):
            seats.extend(self.tables[i].seats)
            i+=1
        random.shuffle(seats)
        
        i=0
        while i < len(seats):
            if not seats[i].free:
                seats[i].pop()
            i+=1
        
        i=0
        while i < len(names):
            seats[i].set_occupant(names[i])
            i+=1
        
    def display(self):
        for i, table in enumerate(self.tables):
            print(f"Table {i+1}:")
            for seat in table.seats:
                occupant = seat.occupant if not seat.free else "Empty"
                print(f"  - {occupant}")
                
    def store(self, filename):
        with open(filename, 'w') as f:
            for i, table in enumerate(self.tables):
                f.write(f"Table {i+1}:\n")
                for seat in table.seats:
                    occupant = seat.occupant if not seat.free else "Empty"
                    f.write(f"  - {occupant}\n")
                    
                    
```
